# Making changes to the data

```python
# add total colum
df['Total'] = df.iloc[:,4:10].sum(axis=1)

# df['Total'] = df['HP'] + df['Attack'] + df['Defence'] ...... + df['Speed']
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head(5)
```

## Filtering Data

```python
new_df = df.loc[(df['Type 1'] == 'Grass' )& (df['Type 2'] == 'Poison') & (df['HP'] > 70) ]

# Or condition
    # df.loc[(df['Type 1'] == 'Grass' ) |(df['Type 2'] == 'Poison')]


# reset the index number
new_df = new_df.reset_index(drop = True)

new_df
```

```python
# str.contains function   not is ~
import re
df.loc[df['Type 1'].str.contains('fire|Grass', flags = re.I, regex= True)].head(10) # no space around |
# flags = re.I   ~ is ignore the case of letter ex) fire or Fire FIRE
```

```python
# start letters are pi     ^ is frst letter    if I do not use ^, I 
df.loc[df['Name'].str.contains('^pi[a-z]', flags = re.I, regex= True)].head(5)
```
